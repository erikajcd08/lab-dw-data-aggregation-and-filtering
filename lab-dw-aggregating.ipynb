```python
import pandas as pd

url='https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv'
df=pd.read_csv(url)

df.head()
```

```python
df.columns= df.columns.str.lower().str.replace(" ","_")
```

```python
df_filtrado= df[(df['total_claim_amount']< 1000) & (df['response']== 'Yes')]
df_filtrado.head()
df_filtrado.shape
```

```python
df.shape
```

El Dataframe original tiene 10910 clientes y el Dataframe filtrado se ha reducido hasta 1399 clientes que tienen una reclamación baja y han respondido "Yes" en la campaña.

```python
df_filtrado_2= df[df['response']== 'Yes']
```

```python
df.columns
```

```python
df_filtrado_2.groupby(['policy_type','gender'])[['monthly_premium_auto','customer_lifetime_value']].mean()
```

En esta última tabla se puede observar que la prima mensual más alta es Personal Auto en género Femenino con un importe de 98.998148. En customer_lifetime_value hay siempre diferencias al alza en las mujeres excepto en Special auto. El segmento más valioso son las mujeres con la póliza Personal auto porque tienen la prima mensual más alta, y generan más valor a largo plazo.

```python
df.groupby(['education','gender'])['customer_lifetime_value'].agg(['max','min','median'])
```

El grupo que tiene el máximo más alto es High School or Below en la sección masculina. El grupo con el mímimo más alto es Doctor en la sección femenina. Dentro de los mismos niveles educativos hay diferencias entre los sexos siendo en la mayoría de ellos, las mujeres con valores más alto que los hombres a excepción de College y High School or Bellow

```python
df['month'] = df['effective_to_date'].dt.month
df[['effective_to_date', 'month']].head()
```

```python
df.pivot_table(index ='state',columns='month',values='effective_to_date',aggfunc=)
```

El estado que vendió más pólizas fue California, y el mes que más se vendió fue el mes 1 con 5005

```python
conteo_polizas= df.groupby(['state','month'])['customer'].count()
conteo_polizas
```

```python
conteo_polizas.sort_values(ascending=False)
```

```python
top3_estados = conteo_polizas.groupby('state').sum().sort_values(ascending=False).head(3).index
top3_estados
```

```python
conteo_polizas[conteo_polizas.index.get_level_values('state').isin(top3_estados)]
```

El estado que lidera más ventas es California en el mes 1 con 1918 pólizas
